## **Gráficos Relacionales**

Los gráficos relacionales son herramientas fundamentales en el análisis de datos que nos permiten explorar y visualizar cómo se relacionan dos o más variables entre sí. Estos gráficos son especialmente útiles para entender las asociaciones y patrones que pueden no ser evidentes al observar los datos en forma tabular. Aquí te explico el contexto y la utilidad de los gráficos relacionales:

#### 1. Objetivo de los Gráficos Relacionales:

Los gráficos relacionales se utilizan para visualizar la relación entre variables, permitiendo una comprensión más profunda de cómo una variable afecta o se relaciona con otra. Son útiles para identificar tendencias, correlaciones, y patrones en los datos. 

#### 2. Tipos de Gráficos Relacionales:

- **Gráfico de Dispersión (Scatter Plot):** Muestra la relación entre dos variables numéricas. Cada punto en el gráfico representa una observación en el conjunto de datos, con la posición en el eje X e Y correspondiente a los valores de las dos variables.
  
- **Gráfico de Líneas (Line Plot):** Ideal para visualizar la relación entre variables cuando una de ellas representa una medida de tiempo. Permite observar tendencias y cambios a lo largo del tiempo.
  
- **Gráfico de Relación (relplot):** Una función de Seaborn que permite crear gráficos de dispersión o de líneas con facilidad. La función `relplot()` es especialmente útil para crear gráficos que muestran cómo una variable se relaciona con otra, con opciones para ajustar el tipo de gráfico según el contexto.

#### 3. Aplicaciones Comunes:

- **Análisis de Tendencias:** Los gráficos de líneas son útiles para analizar cómo las variables cambian a lo largo del tiempo. Por ejemplo, podrías usar un gráfico de líneas para visualizar la evolución del número de pasajeros a bordo del Titanic a lo largo de las distintas fechas de embarque.

- **Detección de Correlaciones:** Los gráficos de dispersión ayudan a identificar la fuerza y dirección de una relación entre dos variables. Por ejemplo, podrías analizar la relación entre la tarifa pagada y la edad de los pasajeros para ver si existe alguna tendencia.

- **Comparación de Subgrupos:** Los gráficos relacionales también permiten comparar diferentes subgrupos dentro de los datos. Por ejemplo, podrías comparar la relación entre la clase socioeconómica y la supervivencia de los pasajeros, usando diferentes colores o marcadores para distinguir entre clases.

#### 4. Ventajas:

- **Visualización Clara:** Ofrecen una representación visual clara de las relaciones entre variables, facilitando la interpretación de datos complejos.
  
- **Identificación de Patrones:** Ayudan a identificar patrones, tendencias y anomalías que pueden no ser evidentes en los datos brutos.

- **Facilidad de Interpretación:** Los gráficos relacionales permiten una interpretación intuitiva de las relaciones entre variables, apoyando el análisis exploratorio de datos.

#### 5. Herramientas:

- **Seaborn:** Una biblioteca de Python que proporciona una interfaz fácil de usar para crear gráficos relacionales con `relplot()`. Permite ajustar el tipo de gráfico y personalizar la visualización según las necesidades del análisis.

En resumen, los gráficos relacionales son esenciales para explorar y entender la interrelación entre variables, facilitando la interpretación de datos y la identificación de patrones importantes en el análisis.

### 3.1. relplot

Interfaz a nivel de figura para dibujar gráficos relacionales en un FacetGrid.

La función `relplot()` recibe este nombre porque está diseñada para visualizar muchas relaciones estadísticas diferentes. Mientras que los gráficos de dispersión son una forma altamente efectiva de hacer esto, las relaciones en las que una variable representa una medida de tiempo se representan mejor mediante una línea. La función `relplot()` tiene un conveniente parámetro `kind` que te permite cambiar fácilmente a esta representación alternativa.


```python
sns.relplot(x="Age", y="Fare", col="Pclass", hue="Sex", style="Sex",kind="line", data=full_data) # scatter can be used instead of "line" plot
plt.show()
```

### 3.2. scatterplot

---

**Scatter plot** es la forma más conveniente de visualizar la distribución donde cada observación se representa en un gráfico bidimensional a través de los ejes x e y.


```python
plt.figure(figsize=(8,8))
sns.scatterplot(x="Age", y="Fare", hue="Sex", data=full_data)
plt.show()
```

### 3.3. lineplot

---

**Lineplot** es una forma de trazar un gráfico de líneas con la posibilidad de varios agrupamientos semánticos.

La relación entre **x** y **y** se puede mostrar para diferentes subconjuntos de los datos utilizando los parámetros **hue**, **size** y **style**. Estos parámetros controlan qué semánticas visuales se utilizan para identificar los diferentes subconjuntos. Es posible mostrar hasta tres dimensiones independientemente utilizando los tres tipos semánticos, pero este estilo de gráfico puede ser difícil de interpretar y a menudo es ineficaz. Utilizar semánticas redundantes (es decir, tanto **hue** como **style** para la misma variable) puede ser útil para hacer los gráficos más accesibles.

El tratamiento predeterminado del **hue** (y, en menor medida, **size**), si está presente, depende de si la variable se infiere que representa datos **“numéricos”** o **“categóricos”**. En particular, las variables numéricas se representan con un mapa de colores secuencial por defecto, y las entradas de la leyenda muestran **“ticks”** regulares con valores que pueden o no existir en los datos. Este comportamiento se puede controlar a través de varios parámetros.

Por defecto, el gráfico agrega sobre múltiples valores de **y** en cada valor de **x** y muestra una estimación de la tendencia central y un intervalo de confianza para esa estimación.


```python
plt.figure(figsize=(8,8))
sns.lineplot(x="Age", y="Fare", hue="Sex", style="Sex", data=full_data)
plt.show()
```

Cuando una o ambas variables en estudio son categóricas, utilizamos gráficos como **`striplot()`, `swarmplot()`, etc.**. Seaborn proporciona una interfaz para hacerlo.

### 4.1. Barplot

---

**`barplot()`** muestra la relación entre una variable categórica y una variable continua. Los datos se representan en barras rectangulares donde la longitud de la barra representa la proporción de los datos en esa categoría. El **Bar plot** representa la estimación de la **tendencia central**.

**Nota:** No confundir **Bar plot** con **Histograma**. Aquí están las principales diferencias entre ellos:

- **Barplot:**
  - **Variables:** Utiliza una variable categórica en el eje x y una variable continua en el eje y.
  - **Propósito:** Muestra la relación entre categorías y su valor promedio o alguna medida de tendencia central.
  - **Ejemplo:** Comparar el promedio de tarifas (`Fare`) pagadas por diferentes clases socioeconómicas (`Pclass`).

- **Histograma:**
  - **Variables:** Utiliza una sola variable continua en el eje x y muestra la frecuencia de ocurrencia en el eje y.
  - **Propósito:** Muestra la distribución de una variable continua, es decir, cómo se distribuyen los valores a lo largo de su rango.
  - **Ejemplo:** Mostrar cómo se distribuyen las edades (`Age`) de los pasajeros.

El **Bar plot** es útil para comparar diferentes categorías, mientras que el **Histograma** es útil para entender la distribución de una sola variable continua.


```python
plt.figure(figsize=(8,8))
sns.barplot(x="Sex", y="Survived", hue="Pclass", data=full_data)
plt.show()
```

### 4.2. countplot

---

Un caso especial en **barplot** es mostrar el número de observaciones en cada categoría en lugar de calcular una estadística para una segunda variable. Para esto, usamos `countplot()`.


```python
sns.countplot(x="Survived", data=full_data, palette="Blues");
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(7, 9), sharex=True, sharey=True)

# Configuración de colores
palette = ["#f9dbbd", "#DA627D", "#450920"]  # Paleta azul para las clases

# Primer gráfico: Número absoluto de pasajeros por Sexo y Clase
sns.countplot(x="Sex", hue="Pclass", data=full_data, palette=palette, ax=axs[0])
axs[0].set_title("Número absoluto de pasajeros por Sexo y Clase")
axs[0].set_xlabel("Sexo")
axs[0].set_ylabel("Pasajeros")
axs[0].legend(title="Clase")

# Segundo gráfico: Número absoluto de sobrevivientes por Sexo y Clase
sns.countplot(x="Sex", hue="Pclass", data=full_data[full_data.Survived==1], palette=palette, ax=axs[1])
axs[1].set_title("Número absoluto de sobrevivientes por Sexo y Clase")
axs[1].set_xlabel("Sexo")
axs[1].set_ylabel("Sobrevivientes")
axs[1].legend(title="Clase")

# Ajuste de diseño para evitar superposiciones
plt.tight_layout()
plt.show()
```

**El gráfico muestra que el número de pasajeros en la tercera clase es mayor que en la primera y segunda clase.**

 **Como podemos ver, la proporición de mujeres que sobrevivieron es mucho mayor a la de los hombres.**

---

Los diagramas de caja suelen tener líneas verticales que se extienden desde las cajas, llamadas **bigotes**. Estos bigotes indican la variabilidad fuera de los cuartiles superior e inferior, por lo que los Boxplots también se conocen como **diagrama de caja y bigotes**. Los valores atípicos en los datos se representan como puntos individuales.

# 5. Gráficos de Regresión

La mayoría de las veces, utilizamos conjuntos de datos que contienen múltiples variables cuantitativas, y el objetivo de un análisis a menudo es relacionar esas variables entre sí. Esto se puede hacer mediante las *líneas de regresión*.

Mientras construimos los modelos de regresión, a menudo verificamos la *multicolinealidad*, donde tenemos que observar la correlación entre todas las combinaciones de variables continuas y tomar las medidas necesarias para eliminar la multicolinealidad si existe.

## 5.1. lmplot


```python
sns.lmplot(x="Age", y="Fare", data=full_data)
plt.show()
```

## 5.2. regplot

`regplot` acepta las variables x e y en una variedad de formatos, incluyendo arreglos simples de numpy, objetos Series de pandas, o como referencias a variables en un DataFrame de pandas.


```python
from scipy.stats import pearsonr

plt.subplots(figsize=(10, 10))
sns.regplot(x="Age", y="Fare", data=full_data)
plt.show()

# Eliminar filas con valores NaN en las columnas 'Age' y 'Fare'
cleaned_data = full_data[['Age', 'Fare']].dropna()
# Calcular el coeficiente de correlación de Pearson
correlation_coefficient, _ = pearsonr(cleaned_data['Age'], cleaned_data['Fare'])
# Imprimir el coeficiente de correlación
print(f"Coeficiente de correlación de Pearson entre Edad y Tarifa: {correlation_coefficient:.2f}")
```

```python
plt.subplots(figsize=(10, 10))
sns.regplot(x="Fare", y="Fare", data=full_data)
plt.show()
```

# 6. Gráficos de Matriz

## 6.1. heatmap

Visualizar datos con **heatmaps** es una excelente manera de realizar un análisis exploratorio de datos cuando se tiene un conjunto de datos con múltiples variables. Los heatmaps pueden revelar patrones generales en el conjunto de datos de manera instantánea. Y es muy fácil crear heatmaps atractivos con la biblioteca Seaborn en Python.

Ahora vamos a trazar la matriz de correlación de nuestros datos con un heatmap.


```python
# Selecciona solo columnas numéricas
numeric_data = full_data.select_dtypes(include=['number'])

# Calcular la matriz de correlación
correlation_matrix = numeric_data.corr()

# Crear el heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(correlation_matrix, cmap="YlGnBu", annot=True, fmt=".2f")
plt.title("Matriz de Correlación")
plt.show()
```

**Observamos que las correlaciones entre las variables son débiles.**

# 7. Multi-plot grids

## 7.1. Facet grids

Una técnica útil para explorar los datos a través de varias dimensiones es dibujar múltiples instancias del mismo gráfico en diferentes subconjuntos de tu conjunto de datos. Esta técnica se denomina comúnmente *“rejilla”* o *“paneles”*, y está relacionada con la idea de *“múltiples pequeños”*. Para utilizar estas funciones, tus datos deben estar en un DataFrame de Pandas.

**La rejilla de paneles** forma una matriz de paneles definidos por filas y columnas al dividir las variables. Debido a los paneles, un solo gráfico parece ser múltiples gráficos. Es muy útil para analizar todas las combinaciones en dos variables discretas.

El objeto **FacetGrid** toma un DataFrame como entrada y los nombres de las variables que formarán las dimensiones de fila, columna o tono de la rejilla. Las variables deben ser categóricas, y los datos en cada nivel de la variable se utilizarán para un panel a lo largo de ese eje.

La ventaja de usar Facet es que podemos ingresar otra variable en el gráfico. Podemos crear muchos paneles de columna y alinearlos con las filas de la rejilla.


#### FacetGrid.map

El enfoque principal para visualizar datos en esta cuadrícula es con el método `FacetGrid.map()`

```python
# initialize the FacetGrid object
g = sns.FacetGrid(full_data, col='Survived', row='Pclass')

g.map(plt.hist, 'Age')
g.add_legend()
plt.show()
```

## Diagramas de torta

Para entender mejor la composición de los pasajeros en el conjunto de datos, se puede utilizar un gráfico de torta (diagrama circular). Este gráfico visualiza la proporción de cada categoría de género dentro del total de pasajeros.

### Reemplazo de Valores Nulos en la Columna 'Age' con la Moda

El siguiente código calcula la moda de la columna `'Age'` en un DataFrame y utiliza este valor para reemplazar los valores nulos en la misma columna. Primero, se calcula la moda de la columna `'Age'` utilizando el método `mode()`, que devuelve una Serie con la moda de la columna. Se selecciona el primer valor de esta Serie (el valor más frecuente) usando `[0]`. Luego, se utiliza el método `fillna()` para reemplazar los valores nulos en la columna `'Age'` con la moda calculada. El parámetro `inplace=True` asegura que la operación se realice directamente en el DataFrame sin necesidad de asignar el resultado a una nueva variable.

```python
# Calcular la moda de la columna 'Age' por cada categoría de 'Sex'
age_mode_male = df_0[df_0['Sex'] == 'male']['Age'].mode()[0]
age_mode_female = df_0[df_0['Sex'] == 'female']['Age'].mode()[0]

# Crear una función para aplicar la moda correcta según el sexo
def fill_age_by_sex(row):
    if pd.isnull(row['Age']):
        if row['Sex'] == 'male':
            return age_mode_male
        elif row['Sex'] == 'female':
            return age_mode_female
    else:
        return row['Age']

# Aplicar la función para rellenar los valores faltantes en 'Age'
df_0['Age'] = df_0.apply(fill_age_by_sex, axis=1)
```

### **Evaluación de Métricas: Precision y Recall**

En el contexto de la clasificación binaria, como el análisis del conjunto de datos del Titanic, es crucial evaluar cómo de bien está funcionando nuestro modelo al identificar las instancias correctas. En este ejercicio, estamos particularmente interesados en la métrica de **Recall**. 

**Recall** es una métrica que indica qué tan bien nuestro modelo está capturando todas las instancias de la clase positiva. En el caso del Titanic, esta clase positiva representa a los pasajeros que sobrevivieron. Optimizar el **Recall** es importante porque queremos identificar el mayor número posible de pasajeros que sobrevivieron, lo cual es fundamental para evaluar cómo el modelo puede mejorar la identificación de sobrevivientes.

A continuación, se detallan dos métricas importantes:

#### **Precision**

<figure>
<center>
<img src='https://raw.githubusercontent.com/jmpaiba123/Bootcamp_analitica/main/Data/Images/matriz_confusion.png' width="400" height="350" />

* **Precision:** ¿De las predicciones positivas de mi algoritmo, cuántas son realmente positivas? En el contexto del Titanic, ¿de todas las veces que el modelo predice que un pasajero sobrevivirá, cuántos realmente sobreviven?

$$ Precision =  \frac{TP}{TP+FP}$$

$$ Precision =  \frac{80}{80+10}$$

$$ Precision =  0.89$$

Precision se enfoca en la exactitud de las predicciones positivas del modelo, asegurando que las predicciones de supervivencia sean correctas. Por ejemplo, si el modelo predice que 90 pasajeros sobrevivirán y 80 realmente sobreviven (Verdaderos Positivos, TP), pero 10 no sobreviven (Falsos Positivos, FP), la precisión es del 89%.

* **Recall (Sensibilidad):** ¿De todos los casos actuales, cuántos logra identificar mi algoritmo? De todos los pasajeros que realmente sobrevivieron, ¿cuántos logra identificar correctamente mi modelo?

$$ Recall =  \frac{TP}{TP+FN}$$

$$ Recall =  \frac{80}{80+20}$$

$$ Recall =  0.80$$

Recall mide qué tan bien el modelo detecta a los pasajeros que realmente sobrevivieron. Un alto Recall significa que el modelo está capturando la mayoría de los pasajeros sobrevivientes, aunque podría incluir algunos que no lo son (más falsos positivos). Por ejemplo, si en realidad 100 pasajeros sobrevivieron y el modelo identifica correctamente a 80 de ellos (Verdaderos Positivos, TP), pero no identifica a 20 que también sobrevivieron (Falsos Negativos, FN), el Recall sería del 80%.

En nuestro análisis del Titanic, nos enfocamos en optimizar el **Recall** para asegurarnos de identificar la mayor cantidad posible de pasajeros que sobrevivieron. Esto es especialmente importante en escenarios donde queremos minimizar el riesgo de pasar por alto a los pasajeros que sobrevivieron.

# **Árboles de Decisión en Problemas de Clasificación Binaria**

<figure>
<center>

Un **árbol de decisión** es un modelo de aprendizaje automático utilizado para tareas de clasificación y regresión. En problemas de **clasificación binaria**, se usa para predecir a qué una de dos clases pertenece una muestra.

## ¿Cómo Funciona?

1. **División de Datos**: El árbol de decisión divide los datos en diferentes nodos basados en características específicas. Cada nodo interno representa una pregunta o prueba sobre una característica, y cada rama representa el resultado de la prueba.

2. **Nodos**:
   - **Nodos Internos**: Representan pruebas en una característica. Cada nodo divide los datos en función del valor de una característica.
   - **Nodos de Hoja**: Representan las clases finales en las que se clasifica una muestra. Estos nodos contienen la predicción final del modelo.

3. **Criterios de Divisiones**:
   - **Impureza de Gini**: Mide cuán mezcladas están las clases en el nodo. Un nodo con baja impureza de Gini es más puro.
   - **Entropía**: Mide la cantidad de incertidumbre o desorden en los datos. La entropía baja indica un nodo más puro.

## Ventajas
- **Interpretabilidad**: Los árboles de decisión son fáciles de interpretar y visualizar.
- **No Requiere Normalización de Datos**: No es necesario normalizar o estandarizar los datos.

## Desventajas
- **Sobreajuste**: Los árboles de decisión pueden sobreajustar los datos de entrenamiento si no se podan adecuadamente.
- **Sensibilidad a Pequeñas Variaciones**: Pueden ser inestables con pequeñas variaciones en los datos.

Un árbol de decisión para clasificación binaria puede ser una herramienta poderosa para predecir la pertenencia a dos clases, proporcionando decisiones claras basadas en las características de los datos.


## **Entrenamiendo del Modelo**

**Gini:** Esto representa la impureza de Gini del nodo. Mide cuán mezcladas están las clases en el nodo. Un valor más bajo de Gini indica un nodo más puro, lo que significa que la mayoría de las muestras pertenecen a una sola clase.

**Muestras:** Este es el número de muestras (puntos de datos) que llegan a este nodo. Te dice cuántos puntos de datos se usan para tomar la decisión en este nodo.

**Valor:** Muestra el número de muestras para cada clase dentro del nodo. Generalmente se muestra como un arreglo donde cada elemento representa el conteo de muestras para cada clase. Para clasificación binaria, podrías ver algo como [100, 50], donde 100 es el conteo para la clase 0 y 50 es para la clase 1.

**Clase:** Esta es la clase a la que se clasifica el nodo. Para cada nodo, muestra la clase con el mayor número de muestras. En la clasificación binaria, es la clase con la mayoría de las muestras en ese nodo.

```python
def plot_confusion_matrices_and_reports(y_train_pred, y_test_pred, y_train_true, y_test_true):
    
    # Evaluation for the training set
    train_accuracy = accuracy_score(y_train_true, y_train_pred)
    train_auc = roc_auc_score(y_train_true, y_train_pred)
    train_report = classification_report(y_train_true, y_train_pred)
    train_conf_matrix = confusion_matrix(y_train_true, y_train_pred)

    # Evaluation for the test set
    test_accuracy = accuracy_score(y_test_true, y_test_pred)
    test_auc = roc_auc_score(y_test_true, y_test_pred)
    test_report = classification_report(y_test_true, y_test_pred)
    test_conf_matrix = confusion_matrix(y_test_true, y_test_pred)

    def plot_confusion_matrix(cm, report, title='Confusion Matrix', cmap=plt.cm.Blues):
        fig, axes = plt.subplots(1, 2, figsize=(8, 4))

        # Plot confusion matrix
        sns.heatmap(cm, annot=True, fmt='d', cmap=cmap, 
                    xticklabels=['Class 0', 'Class 1'], 
                    yticklabels=['Class 0', 'Class 1'], ax=axes[0])
        axes[0].set_title(title)
        axes[0].set_xlabel('Predicted')
        axes[0].set_ylabel('Actual')

        # Plot classification report
        axes[1].text(0.5, 0.5, report, ha='center', va='center', fontsize=10, 
                     bbox=dict(facecolor='white', alpha=0.7), multialignment='right')
        axes[1].set_title('Métricas de Clasificación')
        axes[1].axis('off')  # Turn off the axis

        plt.tight_layout()
        plt.show()

    # Plot for training set
    plot_confusion_matrix(train_conf_matrix, train_report, title='Conjunto Entrenamiento', cmap=plt.cm.Blues)

    # Plot for test set
    plot_confusion_matrix(test_conf_matrix, test_report, title='Conjunto de Prueba', cmap=plt.cm.Greens)
```

```python
# Comparando resultados en conjunto de Entrenamiento vs Prueba
plot_confusion_matrices_and_reports(y_pred_train_dt, y_pred_test_dt, y_train, y_test)    
```

```python
def compute_metrics(y_train_true, y_train_pred, y_test_true, y_test_pred, y_train_pred_proba):
    
    # Accuracy for training and test set
    accuracy_train = round(100.0 * accuracy_score(y_train_true, y_train_pred), 2)
    accuracy_test = round(100.0 * accuracy_score(y_test_true, y_test_pred), 2)
    
    # AUC for the training set
    #auc = round(roc_auc_score(y_train_true, y_train_pred_proba), 3)
    
    # Compute the classification report
    report = classification_report(y_test_true, y_test_pred, output_dict=True)
    
    # Extract precision, recall, F1 score for class 1
    precision_class_1 = round(report['1']['precision'], 3)
    recall_class_1 = round(report['1']['recall'], 3)
    f1_class_1 = round(report['1']['f1-score'], 3)
    
    return {
        'accuracy_train': accuracy_train,
        'accuracy_test': accuracy_test,
       # 'auc': auc,
        'precision_class_1': precision_class_1,
        'recall_class_1': recall_class_1,
        'f1_class_1': f1_class_1
    }
```

```python
metrics_dt = compute_metrics(y_train, y_pred_train_dt, y_test, y_pred_test_dt, y_prob_train_dt)
metrics_dt
```

La **regresión logística** es un modelo estadístico utilizado para predecir la probabilidad de una variable dependiente categórica, generalmente binaria. A diferencia de la regresión lineal, que se utiliza para variables continuas, la regresión logística estima la probabilidad de que un evento ocurra mediante una función logística, que transforma la salida lineal del modelo en una probabilidad entre 0 y 1. Esta técnica es ampliamente utilizada en problemas de clasificación, como la detección de fraude, diagnóstico médico y análisis de mercado.

En el contexto de aprendizaje automático, la regresión logística es valorada por su simplicidad y eficacia. Aunque se basa en un supuesto de relación lineal entre las características y la probabilidad de la clase positiva, su capacidad para manejar variables independientes e interpretar la importancia de las características la hace útil en una variedad de aplicaciones. La salida del modelo puede ser interpretada como una probabilidad, lo que facilita la toma de decisiones basadas en umbrales de probabilidad ajustables.

Un Random Forest es un modelo de aprendizaje automático que usa un grupo de árboles de decisión para hacer predicciones.

**¿Cómo funciona?**

1. Árboles de Decisión: Un árbol de decisión toma decisiones de forma jerárquica, dividiendo los datos en diferentes categorías basadas en características.

2. Bosque Aleatorio: En lugar de un solo árbol, Random Forest crea muchos árboles de decisión (un "bosque") y los combina para hacer una predicción más precisa y estable.

3. Promedio de Resultados: Para la clasificación, el modelo vota por la categoría más común entre los árboles. Para la regresión, promedia las predicciones de todos los árboles.

Ventajas

* Menos Riesgo de Sobreajuste: Combina múltiples árboles, lo que reduce el riesgo de que el modelo se ajuste demasiado a los datos de entrenamiento.
* Precisión Mejorada: Generalmente, proporciona predicciones más precisas que un solo árbol de decisión.
* En resumen, Random Forest es como pedir la opinión de muchos expertos (árboles de decisión) y tomar una decisión basada en la mayoría.
